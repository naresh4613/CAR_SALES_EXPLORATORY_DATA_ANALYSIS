```python
car_sales_copy["drive"]=car_sales_copy["drive"].fillna("front")
```

```python
car_sales.isnull().sum()
```

```python
print(car_sales_copy.duplicated().sum())
```

```python
car_sales_copy.loc[car_sales.duplicated(keep=False), :]
```

```python
car_sales_copy['engV']=car_sales_copy.groupby(['car','body'])['engV'].transform(lambda x: x.fillna(x.median()))
```

```python
car_sales_copy.isnull().sum()
```

422 missing value of engV has been replaced however , still 10 entries are left as missing,lets see the missing value data.

```python
car_sales_copy[car_sales_copy.engV.isnull()]
```

Replacing NaN  values of drive with most common values of drive from car and body group.

```python
def f(x):
    if x.count()<=0:
        return np.nan
    return x.value_counts().index[0]

car_sales_copy['drive'] = car_sales_copy['drive'].fillna(car_sales_copy.groupby(['car','body'])['drive'].transform(f))
#CarSales_Data[CarSales_Data.drive.isnull()]
```

```python
car_sales_copy[car_sales_copy.drive.isnull()]
```

```python
car_sales_copy.isnull().sum()
```

```python
car_sales_copy.dropna(subset=['engV'],inplace=True)
car_sales_copy.dropna(subset=['drive'],inplace=True)
car_sales_copy.isnull().sum()
```

Dropping entries with price <=0

```python
carsales_copy = car_sales_copy.drop(car_sales_copy[car_sales_copy.price <= 0 ].index)
```

```python
car_sales_copy.price[car_sales_copy.price ==0].count()
```

```python
b = car_sales_copy["mileage"].median()
car_sales_copy["mileage"]=car_sales_copy["mileage"].replace(0,b)
```

```python
car_sales_copy[car_sales_copy.mileage == 0]
```

```python
car_sales_copy
```

```python
profile_cleaned = pandas_profiling.ProfileReport(car_sales_copy)
profile_cleaned.to_file(output_file="car_sales_post_preprocessing.html")
```
